```python
# numpy = numerical routines in python.
import numpy as np
# IMPLICIT: import matplotlib.pylab as plt
```

```python
# I remember there is a routine in numpy to sample 
# from a normal distribution but I'm fuzzy on the details

# start typing - type np.ran , then hit TAB
np.ran
```

```python
# ok so it's called np.random.normal
# but what arguments does it expect?
# type np.random.normal( ), put the cursor between the () and hit shfit-TAB
np.random.normal()
```

```python
# ok, loc --> mean
#     scale --> stdev
#     size --> how many values to sample
```

```python
# let's make up a 'dataset'
x = np.random.normal( loc=0, scale=10, size=100 )
y = np.random.normal( loc=20, scale=2, size=100 )
z = 3*y + x
```

```python
# let's look at some of these
print 'x: ', x[0:10]
print ' '
print 'y: ',y[0:10]

# makes sense - X ~ N(0,100), Y ~ N(20,4)
```

```python
# numpy as lots of routines to manipulate arrays, e.g., taking mean
print 'x mean: ',np.mean(x)
print 'x std: ',np.std(x)
print '  '
print 'y mean: ',np.mean(y)
print 'y std: ',np.std(y)
print '  '
print 'z mean: ',np.mean(z)
print 'z std: ',np.std(z)
```

```python
# first plot: scatterplot
plt.scatter( x, y )
plt.title('x versus y')
plt.xlabel(' X ~ N(0,100)')
plt.ylabel(' Y ~ N(20,4)')
```

```python
# first plot: scatterplot
plt.scatter( x, z )
plt.title('x versus z')
plt.xlabel(' X ~ N(0,100); Y ~ N(20,4)')
plt.ylabel(' Z = 3*Y+X ')
```

```python
# you can change color
plt.scatter( x, z, color='red' )
plt.title('x versus z')
plt.xlabel(' X ~ N(0,100); Y ~ N(20,4)')
plt.ylabel(' Z = 3*Y+X ')
```

```python
# you can change the marker style
#  -->  http://matplotlib.org/api/markers_api.html#module-matplotlib.markers
plt.scatter( x, z, marker='x', color='blue' )
plt.title('x versus z')
plt.xlabel(' X ~ N(0,100); Y ~ N(20,4)')
plt.ylabel(' Z = 3*Y+X ')
```

```python
# multiple calls on one plot
plt.scatter( x, z, marker='x', color='blue', label='Z versus X' )
plt.scatter( x, y, marker='o', color='green', label='Y versus X' )
plt.legend()
```

```python
# adjacent plots
my_figure, my_axes = plt.subplots( 1, 2, sharey=True, sharex=True )
my_axes[0].scatter( x, y, marker='x', color='green', label='Y versus X' )
my_axes[1].scatter( x, z, marker='x', color='blue', label='Z versus X' )

my_axes[0].legend()
my_axes[1].legend()

# what happens if you turn off sharex or sharey (set to =False)?
```

```python
# line plot
# let's make a sorted version of the Y values
ysorted = np.sort(y)
# let's make an array of 0....|Y|  (i.e., ranks)
a = np.arange(0, y.shape[0])
```

```python
plt.plot( a, ysorted )
# what happens if you has passed in y instead of ysorted?
```

```python
# box plots
_ = plt.boxplot( x=[ x,y,z ], labels=['x','y','z'] )
```

```python
# histogram of many stripes
_ = plt.hist( x=z,bins=100 )
plt.figure()
```

```python
_ = plt.hist( x=z,bins=100, edgecolor='none' )
plt.figure()
```

```python
_ = plt.hist( x=z, histtype='step', cumulative=True )
plt.figure()
_ = plt.hist( x=z, histtype='step', normed=True, cumulative=True )
plt.figure()
# narrower bins
_ = plt.hist( x=z, bins=100, histtype='step', normed=True, cumulative=True )
plt.figure()
# custom bins
_ = plt.hist( x=z, bins=[0,20,30,31,32,33,34,35,40,50,60,65,80,100], histtype='step', normed=True, cumulative=True )
plt.figure()

```
